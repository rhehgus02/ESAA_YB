# **1. 단순 선형 회귀 분석**
- 전복의 나이를 예측하는 선형회귀모델을 생성하세요.
- 전복의 ‘성별’, ‘키’, ‘지름’, ‘높이’, ‘전체무게’, ‘몸통무게’, ‘내장무게’, ‘껍질무게’를 이용해 ‘껍질의 고리 수’를 예측한 뒤, **예측된 ‘껍질의 고리 수’에 1.5를 더하면 전복의 나이**가 됩니다.

```python
# 기본 모듈 불러오기
import numpy as np
import pandas as pd
```

**1) 데이터 load 및 변형** 

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 데이터 로드
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ESAA/data/abalone.csv")
print(data.shape)
data.head()
```

```python
# 성별 M은 Male, F는 Female, I는 Infant 이므로 따로 열 만들기
for label in "MFI":
    data[label] = data["Sex"] == label
data.drop('Sex', axis=1, inplace=True)
data.head()
```

**2) X, y 선택**
: y는 Rings열, X는 Rings열을 제외한 나머지를 선택하되 전부 실수가 되도록 한다.

```python
# X,y 데이터 선택
y = data['Rings']
X = data.drop('Rings', axis=1)
X = X.astype(float)
```

```python
X.head(3)
```

 **3) train/test set 분리**

```python
# 필요한 모듈 불러오기
from sklearn.model_selection import train_test_split
```

```python
# train과 test set 분리 (train:test = 7:3 비율로)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

**4) 선형회귀모델 생성, 모델 예측치 구하기**

```python
# 필요한 모듈 불러오기
from sklearn.linear_model import LinearRegression
```

```python
# 선형회귀모델 생성 및 훈련
mymodel = LinearRegression()
mymodel.fit(X_train, y_train)
```

```python
# 모델 예측치 구하기
y_pred = mymodel.predict(X_test)

# 모델 예측치를 활용해 최종적으로 전복의 나이를 예측
y_pred = y_pred + 1.5  
```

**5) 모델 평가: MSE, RMSE, R2 score, corr 구하기**

```python
# 필요한 모듈 불러오기
from sklearn.metrics import mean_squared_error, r2_score
```

- MSE, RMSE 

```python
# mse, rmse
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f'MSE: {mse:.4f}, RMSE: {rmse:.4f}')
```

- R2 score

```python
#R2 score 측정
R2_score = r2_score(y_test, y_pred)
print(f'R2 score: {R2_score:.4f}')
```

- 회귀 절편값

```python
# 회귀 절편 값
np.round(mymodel.intercept_, 2)
```

- 회귀 계수 값

```python
# 회귀 계수 값
np.round(mymodel.coef_, 2)
```

# **2. Polynomial features**

```python
# PolynomialFeatures 라이브러리 호출
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 임의 데이터 생성

X = np.arange(6).reshape(3, 2)

df = pd.DataFrame(X)
df.columns = ['x_1','x_2']
df
```

```python
# 차원은 2로 설정
# fit_transform 메소드를 통해 데이터 변환
# PolynomialFeatures로 변환 된 데이터를 데이터 프레임 형태로 변환

poly = PolynomialFeatures(degree=2)
df2 = poly.fit_transform(df)
df2 = pd.DataFrame(df2) # PolynomialFeatures로 변환 된 데이터를 데이터 프레임 형태로 변환
df2
```

```python
# df_poly의 컬럼을 1,x1,x2,x1^2,x1*x2,x2^2 로 변경
column_names = ['1','x1','x2','x1^2','x1*x2','x2^2']
df2.columns = column_names
df2
```
