#### 2-2) 다변량 그래프 그리기

* 다변량 그래프: 3개 이상의 변수를 사용한 그래프

* 산점도 그래프에서 점의 색상을 다르게 하거나 점의 크기를 다르게 해서 표현

```python
def recode_sex(sex):
    if sex=='Female':
        return 0
    else :
        return 1
```

>> 성별의 문자열을 정수로 치환해 산점도 그래프에서 색상으로 구분할 수 있도록 함

```python
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips.head()
```

>> recode_sex 함수가 반환한 값(0, 1)을 데이터프레임에 sex_color 열을 새로 생성하여 추가

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,
    c = tips['sex_color'],
    alpha = 0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

>> `s`: 점의 크기, 

>> `c`: 점의 색상

>> `alpha`: 점의 투명도

## **3. seaborn 라이브러리 자유자재로 사용하기**

* seaborn 라이브러리를 사용하면 matplotlib보다 좀 더 화려한 그래프를 그릴 수 있음



#### 3-1) 단변량 그래프 그리기 - 히스토그램 

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 히스토그램 + 밀집도 그래프
ax = plt.subplots( )
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

>> 먼저 `plt.subplots()` 메서드로 기본 틀을 만듦

#### 3-2) 다양한 종류의 이변량 그래프 그리기

```python
# seaborn 라이브러리로 산점도 그래프 그리기(regplot)
ax = plt.subplots( )
ax = sns.regplot(x='total_bill', y='tip', data=tips) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

>> `sns.regplot` 메서드를 사용하면 산점도 그래프와 회귀선을 함께 그릴 수 있음

```python
ax = plt.subplots( )
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False) 
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Tip')
```

>> 회귀선을 제거하려면 `fit_reg=False` 사용

```python
# 산점도 그래프 + 히스토그램(joinplot)
joint = sns.jointplot(x='total_bill', y='tip', data=tips) 
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip') 
joint.fig.suptitle('Joint Plot of Total Bill and Tip', 
                   fontsize=10, y=1.03)
```

>> `x`, `y` 인자에 원하는 열 이름을 지정하고 `data` 인자값으로 데이터프레임을 지정하면 그래프를 그릴 수 있음

```python
# 육각 그래프
hexbin = sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', 
                    fontsize=10, y=1.03)
```

>> 산점도 그래프는 점이 겹쳐 보일 경우 구분하기 어려우므로 이를 해결하기 위해서 육각 그래프를 사용

>> 육각 그래프: 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프

>> 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현됨

>> `jointplot` 메서드에서 `kind="hex"` 사용

```python
# 이차원 밀집도(kdeplot)
ax = plt.subplots()
ax = sns.kdeplot(x=tips['total_bill'], y=tips['tip'], shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

>> `kdeplot` 메서드에 두 개의 열 데이터를 전달하여 이차원 밀집도를 그림

>> `shade=True`를 사용하면 그래프에 음영 효과를 줄 수 있음

```python
# 바그래프(barplot)
ax = plt.subplots( )
ax = sns.barplot(x='time', y='total_bill', data=tips) 
ax.set_title('Bar plot of average total bill for time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Average total bill')
```

>> 바그래프: 지정한 변수의 평균을 계산하여 그림

>> 시간에 따른 지불한 비용의 평균을 나타낸 것

```python
# 박스그래프(boxplot)
ax = plt.subplots( )
ax = sns.boxplot(x='time', y='total_bill', data=tips) 
ax.set_title('Boxplot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

>> 박스 그래프: 최솟값, 1분위수, 중간값, 3분위수, 최댓값, 이상치 등 다양한 통계량을 한 번에 표현하는 그래프

```python
# 바이올린 그래프(violonplot)
ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', data=tips) 
ax.set_title('Violin plot of total bill by time of day') 
ax.set_xlabel('Time of day') 
ax.set_ylabel('Total Bill')
```

#### 3-3) 관계 그래프 그리기

* 관계 그래프: 지금까지 설명한 그래프를 종합한 그래프

* `map_upper` : 대각선 기준으로 위쪽에 그릴 그래프 지정

* `map_lower` : 대각선 기준으로 아래쪽에 그릴 그래프 지정

* `map_diag` : 대각선을 중심으로 그래프 그림

```python
# 관계그래프(pairplot)
fig = sns.pairplot(tips)
```

>> `pairplot` 메서드에 데이터프레임을 넣는 방법으로 간단하게 그릴 수 있음

>> 이때 관계 그래프는 대각선을 기준으로 중복된다는 단점이 있음

```python
# 원하는 그래프로 교체(PairGrid)
pair_grid = sns.PairGrid(tips) 
pair_grid = pair_grid.map_upper(sns.regplot)  # 대각선 위
pair_grid = pair_grid.map_lower(sns.kdeplot)  # 대각선 아래
pair_grid = pair_grid.map_diag(sns.distplot, rug=True) # 대각선
plt.show()
```

>> 대각선 위: 산점도

>> 대각선 아래: 이차원 밀집도

>> 대각선: 히스토그램, 밀집도

#### 3-5) FacetGrid 클래스로 그룹별 그래프 그리기

* `lmplot` 메서드를 사용하는 방법이 복잡하다면 `FacetGrid` 클래스 사용 가능

```python
# time 그룹별 히스토그램, 밀집도, 양탄자 그래프
facet = sns.FacetGrid(tips, col='time') 
facet.map(sns.distplot, 'total_bill', rug=True)
```

>> 먼저 `FacetGrid`에 데이터프레임과 그룹을 구분할 열 이름을 전달

>> `map` 메서드에 그래프의 종류와 x축으로 지정할 열 이름을 순서대로 전달

```python
# day 그룹별 다변량 그래프
facet = sns.FacetGrid(tips, col='day', hue='sex') 
facet = facet.map(plt.scatter, 'total_bill', 'tip') 
facet = facet.add_legend( )
```

>> 'tips' 데이터 집합의 'day' 열로 그룹을 구분

>> 산점도 그래프를 그리기 위해 `scatter`를 첫 번째 인자에 전달

```python
# 두 개의 변수 그룹별 산점도 그래프
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex') 
facet.map(plt.scatter, 'total_bill', 'tip')
```

## **5. seaborn 라이브러리로 그래프 스타일 설정하기**

* `set_style` 메서드를 사용하여 그래프의 스타일을 바꿀 수 있음

* `set_style`은 스타일을 적용할 부분에서 한 번만 실행하면 됨

* 그러면 실행한 이후의 그래프가 모두 지정한 스타일로 바뀜

```python
fig, ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
# 그래프 배경에 가로줄 추가(whitegrid)
sns.set_style('whitegrid') 
fig, ax = plt.subplots( )
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
# for문을 이용하여 모든 스타일을 하나씩 적용
fig = plt.figure( )

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2, 3, plot_position)
    violin = sns.violinplot(x='time' ,y='total_bill', data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
